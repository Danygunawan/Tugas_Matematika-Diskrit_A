Case: Mencari Asymptot di 2 nilainya berapa, dengan persamaan yang sudah diketahui:

1 + 1/2 + 1/3 +...+ 1/n < 2

```python
import matplotlib.pyplot as plt

def harmonic_number(n):
    return sum(1.0/i for i in range(1, n+1))

target = 2
N = 30

print("Pernyataan: 1 + 1/2 + 1/3 + ... + 1/n < 2\n")

for n in range(1, N+1):
    Hn = harmonic_number(n)
    condition = Hn < target
    print(f"n={n}: Hn={Hn:.6f}, Pernyataan {condition}")

print("\n=================================")
print("Pernyataan benar (True) untuk n = [1, 2, 3]")
print("Pernyataan salah (False) mulai di n = 4 dan seterusnya")

values = [harmonic_number(i) for i in range(1, N+1)]

plt.figure(figsize=(8,6))
plt.plot(range(1, N+1), values, 'bo-', label="Harmonic Series")
plt.axhline(y=2, color='r', linestyle='--', label="y = 2")

first_fail = next(i for i,v in enumerate(values, start=1) if v >= target)
plt.scatter(first_fail, values[first_fail-1], color='green', s=100, zorder=5,
            label=f"n={first_fail}, Hn={values[first_fail-1]:.3f}")

plt.title("Visualisasi Deret Harmonik vs y=2")
plt.xlabel("n")
plt.ylabel("H(n)")
plt.legend()
plt.grid(True)
plt.show()
```
